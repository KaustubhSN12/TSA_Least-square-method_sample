##created a sample data of a company

```python
import pandas as pd

# Creating the 24-month dataset
data = {
    'Month_Number': list(range(1, 25)),
    'Month_Name': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] * 2,
    'Year': [2024] * 12 + [2025] * 12,
    'Sales_Revenue': [
        120000, 125000, 140000, 138000, 150000, 155000, 170000, 165000, 180000, 185000, 200000, 210000,
        205000, 220000, 235000, 230000, 245000, 255000, 270000, 265000, 285000, 290000, 310000, 325000
    ]
}

df = pd.DataFrame(data)

# Save to Excel
df.to_excel('company_sales_24.xlsx', index=False)
print("File 'company_sales_24.xlsx' has been created successfully!")
```

Step 2: The Analysis Code

---


Once the file is created, you can use the code below to solve the "Big Company" problem using the Least Square Method.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the generated data
df = pd.read_excel('company_sales_24.xlsx')

# 2. Least Square Math (Automated)
n = len(df)
midpoint = (n + 1) / 2
df['x'] = 2 * (df['Month_Number'] - midpoint)

# Calculations for a and b
a = df['Sales_Revenue'].mean()
b = np.sum(df['x'] * df['Sales_Revenue']) / np.sum(df['x']**2)

# 3. Generate Trend Line
df['Trend_Value'] = a + (b * df['x'])

# 4. Professional Visualization
plt.figure(figsize=(12, 6))

# Actual Data Line
plt.plot(df['Month_Number'], df['Sales_Revenue'], marker='o', label='Actual Revenue', color='blue', alpha=0.6)

# Trend Line
plt.plot(df['Month_Number'], df['Trend_Value'], label='Statistical Trend Line', color='red', linewidth=3)

# Formatting
plt.title('2-Year Corporate Revenue Trend Analysis', fontsize=14)
plt.xlabel('Month Number (1 to 24)')
plt.ylabel('Revenue (INR)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()

print(f"Trend Equation: y = {a:.2f} + {b:.2f}x")
```

##moving average Solution

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the data (Ensure you ran the previous script to create this file)
df = pd.read_excel('company_sales_24.xlsx')

# 2. Calculate 3-Month Moving Average
# window=3 means it takes the average of 3 consecutive data points
df['Moving_Avg'] = df['Sales_Revenue'].rolling(window=3, center=True).mean()

# 3. Visualization
plt.figure(figsize=(14, 7))

# Plot Actual Sales (The "Zig-Zag" line)
plt.plot(df['Month_Number'], df['Sales_Revenue'], marker='o',
         color='blue', label='Actual Sales', alpha=0.4, linestyle='--')

# Plot Moving Average (The "Smoothed" line)
plt.plot(df['Month_Number'], df['Moving_Avg'], color='green',
         linewidth=3, label='3-Month Moving Average')

# Chart Styling
plt.title('Time Series Smoothing: 3-Month Moving Average', fontsize=16)
plt.xlabel('Month Number')
plt.ylabel('Revenue (INR)')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()

plt.show()

# Display the table to show students the 'NaN' values at the start/end
print(df[['Month_Number', 'Sales_Revenue', 'Moving_Avg']].head(10))
```

###Insights
