```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 1. Setup the data exactly as in the sum
months = np.array([1, 2, 3, 4, 5, 6])
sales = np.array([120, 130, 145, 160, 175, 190])
n = len(months)

# 2. Statistical Transformation (The 'x' column)
# Midpoint is 3.5, multiplied by 2 to get whole numbers
x = 2 * (months - 3.5)

# 3. Calculate Sums (Mirroring the table columns)
sum_y = np.sum(sales)
sum_x = np.sum(x)
sum_xy = np.sum(x * sales)
sum_x_sq = np.sum(x**2)

# 4. Find constants 'a' and 'b'
a = sum_y / n
b = sum_xy / sum_x_sq

# 5. Calculate Trend Values (yt)
trend_values = a + (b * x)

# --- Visualization ---
plt.figure(figsize=(10, 6))

# Plot Actual Data
plt.plot(months, sales, color='blue', marker='o', label='Actual Sales (y)', linewidth=2)

# Plot Trend Line
plt.plot(months, trend_values, color='red', marker='o', linestyle='--', label='Trend Line (yt)')

# Annotate values on the chart
for i, val in enumerate(trend_values):
    plt.text(months[i], trend_values[i] + 2, f'{val:.2f}', color='red', fontweight='bold')

plt.title('Time Series Analysis: Least Square Method', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)

plt.show()

# Print Table for verification
df = pd.DataFrame({
    'Month': months,
    'Sales(y)': sales,
    'x': x,
    'xy': x * sales,
    'x^2': x**2,
    'TrendValue': trend_values
})
print(df)
print(f"\nEquation: y = {a:.2f} + {b:.2f}x")
```
